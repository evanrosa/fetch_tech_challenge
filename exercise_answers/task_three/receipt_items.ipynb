```python
# Find dup receipt items
receipt_items[receipt_items.duplicated(subset=["receipt_id", "barcode", "item_price", "quantity_purchased"], keep=False)]
```

```python
# Are there items without receipts?
receipt_items[~receipt_items["receipt_id"].isin(receipts["receipt_id"])]
```

```python
# Which receipt items need fetch review?
receipt_items[receipt_items["needs_fetch_review"] == True]
```

```python
# Duplicate receipt items that have the same receipt ID and barcode.
# This could indicate duplicate product entries under the same receipt, which might be a data integrity issue.
receipt_items[receipt_items.duplicated(subset=["receipt_id", "barcode"], keep=False)]
```

```python
# Identify receipt items where the final price is less than or equal to zero.
receipt_items[(receipt_items["final_price"] <= 0)]
```

```python
# Find items where price after coupon is greater than item price
receipt_items[receipt_items["price_after_coupon"] > receipt_items["item_price"]]
```

```python
# Find items where final price is greater than item price
receipt_items[receipt_items["final_price"] > receipt_items["item_price"]]
```

```python
# Identify receipt items where the barcode is missing (empty string).
receipt_items[receipt_items["barcode"] == ""]
```

```python
import os
import uuid

# Generate UUID function
def generate_uuid():
    return str(uuid.uuid4())

# Add a new UUID column as Primary Key (PK)
receipt_item_data.insert(0, "receipt_item_id", [generate_uuid() for _ in range(len(receipt_item_data))])

# Export cleaned data
output_dir = "/Users/evro/Documents/code/python/fetch/data/validated"
os.makedirs(output_dir, exist_ok=True)

output_file_pkl = os.path.join(output_dir, "receipt_item_data.pkl")
with open(output_file_pkl, "wb") as f:
    pickle.dump(receipt_item_data, f)
```
